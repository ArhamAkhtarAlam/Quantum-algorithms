```python
# Cell 3: Fourier transform (magnitude) of the values (top-10 output)
import numpy as np

# ensure we have `data` and `show_labels` (prompt if needed)
if 'data' not in globals():
    print('`data` not found. Define or run the previous cell first.')
    data = {}
if 'show_labels' not in globals():
    label_choice = input("Add labels/titles to plots? (y/n): ").strip().lower()
    show_labels = label_choice.startswith('y')

vals = np.array(list(data.values()))
N = vals.size
fft_vals = np.fft.fft(vals)
fft_mag = np.abs(fft_vals)
half = N // 2

# compute positive-frequency magnitudes
pos_mag = fft_mag[:half]

# find top-10 bins by magnitude (descending)
top_n = 10
if top_n > pos_mag.size:
    top_n = pos_mag.size
top_idx = np.argsort(pos_mag)[-top_n:][::-1]  # indices of top magnitudes
top_list = [(int(i), float(pos_mag[i])) for i in top_idx]
print('Top', top_n, 'positive-frequency bins (bin, magnitude):')
for bin_idx, mag in top_list:
    print(f'{bin_idx}: {mag}')

# plot positive-frequency magnitudes
plt.figure(figsize=(14, 4))
plt.plot(pos_mag, linewidth=0.9, color='tab:orange')
plt.scatter(range(half), pos_mag, s=8, color='tab:orange')

# apply labels only if requested
ax = plt.gca()
if show_labels:
    ax.set_xlabel('Frequency bin')
    ax.set_ylabel('Magnitude')
    ax.set_title('FFT magnitude (positive frequencies)')
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.show()
```

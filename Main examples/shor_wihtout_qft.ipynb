```python
# Cell 3: Fourier transform (magnitude) of the values (top-10 output, DC removed)
import numpy as np

# ensure we have `data` and `show_labels` (prompt if needed)
if 'data' not in globals():
    print('`data` not found. Define or run the previous cell first.')
    data = {}
if 'show_labels' not in globals():
    label_choice = input("Add labels/titles to plots? (y/n): ").strip().lower()
    show_labels = label_choice.startswith('y')

vals = np.array(list(data.values()))
N = vals.size
fft_vals = np.fft.fft(vals)
fft_mag = np.abs(fft_vals)
half = N // 2

# remove DC component (bin 0) and consider positive frequencies starting at 1
if half <= 1:
    print('Not enough samples to compute positive-frequency FFT excluding DC.')
    pos_mag = np.array([])
    pos_bins = np.array([])
else:
    pos_mag = fft_mag[1:half]
    pos_bins = np.arange(1, half)

# find top-10 bins by magnitude (descending) excluding DC
top_n = 10
if pos_mag.size == 0:
    print('No positive-frequency bins to report (DC excluded).')
else:
    if top_n > pos_mag.size:
        top_n = pos_mag.size
    top_idx = np.argsort(pos_mag)[-top_n:][::-1]  # indices into pos_mag
    top_list = [(int(pos_bins[idx]), float(pos_mag[idx])) for idx in top_idx]
    print('Top', top_n, 'positive-frequency bins (bin, magnitude)  DC (bin 0) excluded:')
    for bin_idx, mag in top_list:
        print(f'{bin_idx}: {mag}')

# plot positive-frequency magnitudes (DC excluded)
plt.figure(figsize=(14, 4))
if pos_mag.size > 0:
    plt.plot(pos_bins, pos_mag, linewidth=0.9, color='tab:orange')
    plt.scatter(pos_bins, pos_mag, s=8, color='tab:orange')
else:
    plt.text(0.5, 0.5, 'No positive-frequency data (DC excluded)', ha='center')

# apply labels only if requested
ax = plt.gca()
if show_labels:
    ax.set_xlabel('Frequency bin')
    ax.set_ylabel('Magnitude')
    ax.set_title('FFT magnitude (positive frequencies)  DC excluded')
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# Cell 4: Extend the signal, detect where it starts to repeat, and mark it
import numpy as np
import matplotlib.pyplot as plt

# ensure original signal `data` exists
if 'data' not in globals():
    print('`data` not found. Run the signal cell (Cell 2) first.')
    data = {}
vals = np.array(list(data.values()))
N = vals.size

if N == 0:
    print('Signal is empty. Nothing to plot.')
else:
    # compute autocorrelation of zero-mean signal
    x = vals - vals.mean()
    ac_full = np.correlate(x, x, mode='full')
    ac = ac_full[N-1:]  # lags 0..N-1
    ac_norm = ac / ac[0] if ac[0] != 0 else ac
    # find candidate repetition lag: first local maximum above threshold
    lag = None
    threshold = 0.6
    max_search = max(2, N//2)
    for i in range(1, max_search-1):
        if ac_norm[i] > threshold and ac_norm[i] > ac_norm[i-1] and ac_norm[i] >= ac_norm[i+1]:
            lag = i
            break
    if lag is None:
        # fallback: choose the lag with the largest autocorrelation (excluding lag 0)
        if ac_norm[1:].size > 0:
            lag = int(np.argmax(ac_norm[1:]) + 1)
        else:
            lag = 0
    print(f'Detected repetition start lag (index): {lag}   normalized autocorr={ac_norm[lag]:.3f}')

    # ask how many repeats to show
    try:
        rep_in = input('How many times to repeat the signal to extend the plot? (default 3): ').strip()
        repeats = int(rep_in) if rep_in else 3
    except Exception:
        repeats = 3
    if repeats < 1:
        repeats = 1

    # build extended signal by tiling
    extended = np.tile(vals, repeats)

    # plot extended signal and mark repetition start positions
    plt.figure(figsize=(18, 4))
    plt.plot(extended, linewidth=0.9, color='tab:blue')
    plt.scatter(range(len(extended)), extended, s=6, color='tab:blue')

    # mark repetition start (red dashed lines) at lag, lag+N, ...
    for k in range(repeats):
        idx = lag + k * N
        if 0 <= idx < len(extended):
            plt.axvline(idx, color='red', linestyle='--', linewidth=1)

    ax = plt.gca()
    if 'show_labels' in globals() and show_labels:
        ax.set_title('Extended signal with repetition start (red dashed lines)')
        ax.set_xlabel('Index')
        ax.set_ylabel('Value')
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)
    plt.tight_layout()
    plt.show()
```

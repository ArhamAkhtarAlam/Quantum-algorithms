```python
from Crypto.Cipher import PKCS1_OAEP
import ast
from Crypto.PublicKey import RSA # Import RSA to potentially load a key

# Block 3: Your friend's actions (Decryption)




# Get the received ciphertext input from your friend
# This simulates your friend pasting the ciphertext they received from you.
received_ciphertext_string = input("Enter the ciphertext received from your friend (in string format): ")

try:
    # Convert the string representation of the ciphertext back to bytes
    received_ciphertext_bytes = ast.literal_eval(received_ciphertext_string)

    # Use PKCS1_OAEP for decryption with the friend's private key
    cipher_decrypt = PKCS1_OAEP.new(friend_private_key)
    decrypted_message_bytes = cipher_decrypt.decrypt(received_ciphertext_bytes)
    decrypted_message = decrypted_message_bytes.decode('utf-8')

    print("\nMessage decrypted by your friend:")
    print(decrypted_message)

except Exception as e:
    print(f"\nDecryption failed: {e}")
    print("Ensure the correct private key is used for decryption and the ciphertext is valid.")
    print("Also ensure the input ciphertext string is correctly formatted.")
```
